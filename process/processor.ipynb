```python

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = '/usr/share/fonts/opentype/noto/NotoSerifCJK-Regular.ttc'
prop = fm.FontProperties(fname=font_path)
```

```python
df = pd.read_csv('data/ExchangeRates.csv')
df
```

```python
groups = df.groupby('Currency')[['现汇买入', 'Date']]
dfs = {name: group for name, group in groups}
aud = dfs['澳大利亚元']
jpy = dfs['日元']
aud.to_csv('data/aud.csv')
jpy.to_csv('data/jpy.csv')
```

```python
plt.plot(aud['Date'], aud['现汇买入'])
plt.title('澳元现汇买入价格日期走势', fontproperties=prop)
plt.xlabel('澳元现汇买入价', fontproperties=prop)
plt.ylabel('日期（天）', fontproperties=prop)
plt.show()
```

```python
plt.plot(jpy['Date'], jpy['现汇买入'])
plt.title('日元现汇买入价格日期走势', fontproperties=prop)
plt.xlabel('日元现汇买入价', fontproperties=prop)
plt.ylabel('日期（天）', fontproperties=prop)
plt.show()
```
