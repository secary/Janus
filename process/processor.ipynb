```python

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = '/usr/share/fonts/opentype/noto/NotoSerifCJK-Regular.ttc'
prop = fm.FontProperties(fname=font_path)
```

```python
df = pd.read_csv('ExchangeRates.csv')
df
```

```python
groups = df.groupby('Currency')[['现汇卖出', 'Date']]
dfs = {name: group for name, group in groups}
aud = dfs['澳大利亚元'].dropna()
jpy = dfs['日元'].dropna()
aud.to_csv('aud.csv')
jpy.to_csv('jpy.csv')
```

```python
plt.plot(aud['Date'], aud['现汇卖出'])
plt.title('澳元现汇卖出价格日期走势', fontproperties=prop)
plt.xlabel('澳元现汇卖出价', fontproperties=prop)
plt.ylabel('日期（天）', fontproperties=prop)
plt.show()
```

```python
plt.plot(jpy['Date'], jpy['现汇卖出'])
plt.title('日元现汇卖出价格日期走势', fontproperties=prop)
plt.xlabel('日元现汇卖出价', fontproperties=prop)
plt.ylabel('日期（天）', fontproperties=prop)
plt.show()
```
